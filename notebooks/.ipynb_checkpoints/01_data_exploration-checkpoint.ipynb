```python
import os
import pandas as pd
from datasets import load_dataset
```

```python
# Project root = one level above this notebook
PROJECT_ROOT = os.path.abspath("..")

DATA_DIR = os.path.join(PROJECT_ROOT, "data")
RAW_DIR = os.path.join(DATA_DIR, "raw")
PROCESSED_DIR = os.path.join(DATA_DIR, "processed")

os.makedirs(RAW_DIR, exist_ok=True)
os.makedirs(PROCESSED_DIR, exist_ok=True)

print("Project root:", PROJECT_ROOT)
print("Raw data dir:", RAW_DIR)
print("Processed data dir:", PROCESSED_DIR)
```

```python
dataset_name = "btwitssayan/sentiment-analysis-for-mental-health"

# This will download the dataset (train split by default)
hf_ds = load_dataset(dataset_name, split="train")

print(hf_ds)
hf_ds[:3]
```

```python
df = hf_ds.to_pandas()
print(df.head())

output_path = os.path.join(RAW_DIR, "sentiment_mental_health.csv")
df.to_csv(output_path, index=False)

print("Saved raw dataset to:", output_path)
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```
