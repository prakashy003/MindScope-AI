```python
# URL pattern: remove only the URL, keep rest of sentence
URL_PATTERN = re.compile(r"https?://\S+|www\.\S+")

# HTML tags
HTML_TAG_PATTERN = re.compile(r"<.*?>")

# Mentions: @username
MENTION_PATTERN = re.compile(r"@\w+")

# Hashtags: keep the word, drop the '#'
HASHTAG_PATTERN = re.compile(r"#(\w+)")

# Regex for emoji + emoticon keys (we match *exact strings* from EMOJI_EMOTICON_MAP)
emoji_emo_keys = sorted(EMOJI_EMOTICON_MAP.keys(), key=len, reverse=True)
EMOJI_EMOTICON_PATTERN = re.compile("|".join(re.escape(k) for k in emoji_emo_keys))
```

```python
def map_emoji_emoticons(text: str) -> str:
    def repl(m):
        key = m.group(0)
        # For emoticons like ':d', we expect lowercase
        return " " + EMOJI_EMOTICON_MAP.get(key, key) + " "
    return EMOJI_EMOTICON_PATTERN.sub(repl, text)


def clean_text_with_emotions(text: str) -> str:
    if pd.isna(text):
        return ""

    # Convert to string
    text = str(text)

    # 1. Remove URLs (replace them with space, rest of sentence stays)
    text = URL_PATTERN.sub(" ", text)

    # 2. Remove HTML tags if any
    text = HTML_TAG_PATTERN.sub(" ", text)

    # 3. Remove @mentions entirely
    text = MENTION_PATTERN.sub(" ", text)

    # 4. Unwrap hashtags: "#anxiety" -> "anxiety"
    text = HASHTAG_PATTERN.sub(r"\1", text)

    # 5. Lowercase (so emoticons/shortforms standardize: XD -> xd, :D -> :d etc.)
    text = text.lower()

    # 6. Map emojis + emoticons to emotion tags
    text = map_emoji_emoticons(text)

    # 7. Remove any remaining characters except letters, numbers, underscore, and whitespace
    text = re.sub(r"[^a-z0-9_\s]", " ", text)

    # 8. Collapse multiple spaces and trim
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
# Load raw dataset
df = pd.read_csv(RAW_PATH)

print("Columns before dropping index-like columns:", df.columns.tolist())
print("Shape before cleaning:", df.shape)

# 1. Drop index-like columns (e.g. 'Unnamed: 0', 'unnamed: 0')
index_like_cols = [c for c in df.columns if c.lower().startswith("unnamed")]
if index_like_cols:
    print("Dropping index-like columns:", index_like_cols)
    df = df.drop(columns=index_like_cols)
else:
    print("No index-like columns found to drop.")

print("Columns after dropping index-like columns:", df.columns.tolist())

# 2. Drop rows with NaN in text
before_na = df.shape[0]
df = df.dropna(subset=[TEXT_COL])
after_na = df.shape[0]
print(f"Dropped {before_na - after_na} rows with NaN in '{TEXT_COL}'.")

# 3. OVERWRITE TEXT COLUMN WITH CLEANED VERSION
df[TEXT_COL] = df[TEXT_COL].apply(clean_text_with_emotions)

# 4. Drop rows where cleaned text is empty
before_empty = df.shape[0]
df = df[df[TEXT_COL].str.len() > 0]
after_empty = df.shape[0]
print(f"Dropped {before_empty - after_empty} rows with empty cleaned text.")


# 5. Check for duplicates (same text + label)
before_dup = df.shape[0]
df = df.drop_duplicates(subset=[TEXT_COL, LABEL_COL])
after_dup = df.shape[0]
print(f"Dropped {before_dup - after_dup} duplicate rows.")

print("Final shape:", df.shape)

# Quick sanity check
df[[TEXT_COL]].head(10)
```

```python
OUTPUT_PATH = os.path.join(PROCESSED_DIR, "sentiment_mental_health_clean.csv")
df.to_csv(OUTPUT_PATH, index=False)

print("Saved cleaned dataset (with text overwritten) to:")
print(OUTPUT_PATH)
```
